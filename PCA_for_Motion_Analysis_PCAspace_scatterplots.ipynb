```python
files = glob.glob(path + '/*/*.pgm')
```

```python
data = pd.DataFrame()
for path in files:
    img=cv2.imread(path, 0)
    singleImage = pd.Series(img.flatten(),name=path)
    data = data.append(singleImage)
```

```python
fig, axes = plt.subplots(10,10,figsize=(9,9),
     subplot_kw={'xticks':[], 'yticks':[]},
     gridspec_kw=dict(hspace=0.01, wspace=0.01))

for i, ax in enumerate(axes.flat):
     ax.imshow(data.iloc[i].values.reshape(112,92),cmap="gray")
```

```python
from sklearn.decomposition import PCA
#n_components=0.80 means it will return the Eigenvectors that have the 80% of the variation in the dataset
data_pca = PCA(n_components=0.8)
data_pca.fit(data)


fig, axes = plt.subplots(2,10,figsize=(9,3),
 subplot_kw={'xticks':[], 'yticks':[]},
 gridspec_kw=dict(hspace=0.01, wspace=0.01))

for i, ax in enumerate(axes.flat):
     ax.imshow(data_pca.components_[i].reshape(112,92),cmap="gray")
```

```python
var = data_pca.explained_variance_ratio_
var
```

Every Visit

```python
pca_path = r"C:\Users\Daniellab\Documents\Mahad\PCA test case"
hmv_data = pd.read_csv(pca_path + "\heatmap_vectors.csv")
```

```python
hmv_data = hmv_data.set_index('Unnamed: 0')
```

```python
hmv_data.index
```

```python
fig, axes = plt.subplots(6,6,figsize=(9,9),
     subplot_kw={'xticks':[], 'yticks':[]},
     gridspec_kw=dict(hspace=0.01, wspace=0.01))

for i, ax in enumerate(axes.flat):
     ax.imshow(hmv_data.iloc[i].values.reshape(50,50),cmap="gray")
```

```python
from sklearn.decomposition import PCA
hmv_pca = PCA(n_components=4)
hmv_pca_components = hmv_pca.fit_transform(hmv_data)


fig, axes = plt.subplots(2,2,figsize=(9,3),
 subplot_kw={'xticks':[], 'yticks':[]},
 gridspec_kw=dict(hspace=0.01, wspace=0.01))

for i, ax in enumerate(axes.flat):
     ax.imshow(hmv_pca.components_[i].reshape(50,50),cmap="gray")
```

```python
hmv_var = hmv_pca.explained_variance_ratio_
hmv_var
```

```python
hmv_components = hmv_pca.components_
hmv_components.shape
```

```python
hmv_transform = hmv_pca.transform(hmv_data)
projected = hmv_pca.inverse_transform(hmv_transform)
fig, axes = plt.subplots(6,6,figsize=(9,9), subplot_kw={'xticks':[], 'yticks':[]},
            gridspec_kw=dict(hspace=0.01, wspace=0.01))
for i, ax in enumerate(axes.flat):
    ax.imshow(projected[i].reshape(50,50),cmap = 'gray')
plt.savefig(r"C:\Users\Daniellab\Documents\Mahad\PCA test case\PCA space heat map\\EntireVisit_PCAspace.png")
```

```python
hmv_data_components = pd.DataFrame(data = hmv_pca_components, columns = ["PCA1", "PCA2", "PCA3", "PCA4"], index=hmv_data.index)
hmv_data_components 
```

```python
label = ["low"]*19
label.extend(["high"]*19)
label_data = pd.DataFrame(data = label, columns = ["Light_Level"], index=hmv_data.index)
hmv_data_label = pd.concat([hmv_data_components,label_data], axis = 1)
hmv_data_label
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA across all visits', fontsize = 20)

colors = np.linspace(0, 1, 2)
colordict = dict(zip(['low', 'high'], colors))  

label_data["Color"] = label_data["Light_Level"].apply(lambda x: colordict[x])

ax.scatter(hmv_data_label.loc[:, 'PCA1']
               , hmv_data_label.loc[:, 'PCA2']
               , c = label_data.Color
               , s = 50)
plt.savefig(r"C:\Users\Daniellab\Documents\Mahad\PCA test case\PCA space scatterplot\\EntireVisit_PCAspace.png")
```
