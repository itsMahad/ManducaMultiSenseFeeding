```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import glob
import cv2

print("open cv2 version: %s" % cv2.__version__)
```

```python
# read in the annotated list

df = pd.read_csv('./dataFolders/Output/Proboscis/TrackingErrors.csv')
sub_df = df.drop(df[(df.Tracking == 'NO') & (df.WithManualTracking == 'NO')].index)
```

```python
temp = sub_df.name.str.split('_', expand = True)
sub_df['lightLevel'] = temp[0]
sub_df['mothID'] = temp[1] + '_' + temp[2]
sub_df['visit'] = temp[3]
```

```python
sub_df.head()
```

```python
nameOfVideo = []
ss = pd.DataFrame()
for lightLevel, df1 in sub_df.groupby('lightLevel'):
    print(lightLevel)
    for visit, df2 in df1.groupby('visit'):
        print(visit)
        subset= df2[df2.Tracking == 'YES']
        print(subset)
        nameOfVideo.extend(subset.name.values)
        ss = ss.append(subset)
```

```python
ss.to_csv('./dataFolders/Output/Proboscis/CorrectlyTrackedVideos.csv')
```
